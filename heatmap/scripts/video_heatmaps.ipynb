# Video Heatmaps

This notebook attempts to find video heatmaps using different metrics. 

```python
import sys
import librosa
sys.path.insert(0, '/Users/omkarpatil/Documents/event_based_visual_microphone/heatmap/')
import metrics
sys.path.insert(0, '/Users/omkarpatil/Documents/event_based_visual_microphone/utilities/')
import video_frames as vf
import utility_folders as uf
```

```python
video_path = '/Volumes/Omkar 5T/dataset/video_dataset/chips2.mp4'
save_folder = '/Volumes/Omkar 5T/dataset/window_dataset/chips2_'
```

#### **1. First 4 moments of the video**
The code bellow finds the mean, variance, skewness and kurtosis of the video. 

```python
# calculating the moments
mean_left_coor, variance_left_coor, skewness_left_coor, kurtosis_left_coor = metrics.video_moments(video_path)

# displaying the window coordinates and saving the window
print(f"Top left co-ordinate of the mean frame: {mean_left_coor}")
vf.video_window(video_path, save_folder+'mean.mp4', width = mean_left_coor[1], height = mean_left_coor[0], window_dim = [30, 30]) 
print(f"Top left co-ordinate of the variance frame: {variance_left_coor}")
vf.video_window(video_path, save_folder+'variance.mp4', width = variance_left_coor[1], height = variance_left_coor[0], window_dim = [30, 30]) 
print(f"Top left co-ordinate of the skewness frame: {skewness_left_coor}")
vf.video_window(video_path, save_folder+'skewness.mp4', width = skewness_left_coor[1], height = skewness_left_coor[0], window_dim = [30, 30]) 
print(f"Top left co-ordinate of the kurtosis frame: {kurtosis_left_coor}")
vf.video_window(video_path, save_folder+'kurtosis.mp4', width = kurtosis_left_coor[1], height = kurtosis_left_coor[0], window_dim = [30, 30]) 
```

#### **2. Video difference frame**
The code bellow finds the average difference frame between consecutive frames of the video.

```python
# calculating the frame
difference_left_coor = metrics.video_frame_differencing(video_path)

# displaying the window coordinates and saving the window
print(f"Top left co-ordinate of the avg difference frame: {difference_left_coor}")
vf.video_window(video_path, save_folder+'avg_difference.mp4', width = difference_left_coor[1], height = difference_left_coor[0], window_dim = [30, 30]) 
```

#### **3. Video Stability Edge Frame**
The code bellow attempts to find the edges of a video which are most stable.

```python
# calculating the frame
edge_left_coor = metrics.detect_edges_in_video(video_path)

# displaying the window coordinates and saving the window
print(f"Top left co-ordinate of the videos stable edge map: {edge_left_coor}")
vf.video_window(video_path, save_folder+'stable_edge.mp4', width = edge_left_coor[1], height = edge_left_coor[0], window_dim = [30, 30]) 
```

#### **4. Video Entropy Map**
The code bellow attempts to find the entropy map of the video.

```python
# calculating the frame
top_left_coordinate, window_dimension = metrics.process_video_for_entropy(video_path, 16)

# displaying the window coordinates and saving the window
print(f"Top left co-ordinate of the entropy map {top_left_coordinate} and window dimension {window_dimension}")
vf.video_window(video_path, save_folder+'entropy.mp4', width = top_left_coordinate[1], height = edge_left_coor[0], window_dim = [16, 16]) 
```

#### **5. Average video motion frame**
The code bellow uses optical flow equations to find average motion frame of the video.

```python
# calculating the frame
motion_left_coor = metrics.video_temporal_motion(video_path)

# displaying the window coordinates and saving the window
print(f"Top left co-ordinate of the avg motion frame: {motion_left_coor}")
vf.video_window(video_path, save_folder+'avg_motion.mp4', width = motion_left_coor[1], height = motion_left_coor[0], window_dim = [30, 30]) 
```

#### **6. Video Phase Variance**
The code bellow find the variance of the phase frame after performing a FFT and applying a low pass filter.

```python
# calculating the frame
phase_left_coor = metrics.video_phase_variance(video_path)

# displaying the window coordinates and saving the window
print(f"Top left co-ordinate of the phase variance frame: {phase_left_coor}")
vf.video_window(video_path, save_folder+'phase_var.mp4', width = phase_left_coor[1], height = phase_left_coor[0], window_dim = [30, 30]) 
```
