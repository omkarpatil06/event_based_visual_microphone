This script converts a sound array from its file path to a time series plot and spectrogram.

```python
import sys
import soundfile as sf
sys.path.insert(0, '/Users/omkarpatil/Documents/event_based_visual_microphone/utilities/')
import utility_spectrogram as utility
```

The code bellow converts the provided audio array to a spectrogram by first finding its STFT.

The code bellow finds the Segmental SNR of your audio file.

<span style="color:red"> Provide the orignal audiofile path bellow: </span>

```python
orignal_path = '/Volumes/Omkar 5T/dataset/audio_dataset/davis_speech.wav'
```

```python
import librosa

n_fft = 256
hop_length = 128
orig_audio, _ = sf.read(orignal_path)
orig_audio = librosa.resample(orig_audio, orig_sr=44100, target_sr=2200)
orig_audio = orig_audio[:-2400]
utility.show_signal(orig_audio, 'Orignal Marry Had A Little Lamb Speech Chips Bag')
utility.show_spectrogram(orig_audio, audio_samplerate, n_fft, hop_length, 'Original')
utility.segmental_snr(orig_audio, audio)
```
